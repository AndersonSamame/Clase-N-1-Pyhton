### **Ejercicio 1:** Método gráfico de un dato categórico

Analiza de manera gráfica un dato categórico de tu dataset.

```python
#Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('planets')

# Gráfico de barras para el dato categórico 'method'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='method')
plt.title('Métodos de descubrimiento de planetas')
plt.xlabel('Método')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

**La interpretación de tu gráfico anterior:** ------
El método Rdial Velocity es un indicador más facil de describir planetas por el motivo que toda massa está en cosntante movimiento.

### **Ejercicio 2:** Método gráfico de un dato cuantitativo

Analiza de manera gráfica un dato cuantitativo de tu dataset.

```python
#Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('planets')

# Histograma para el dato cuantitativo 'orbital_period'
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='orbital_period', bins=30, kde=True)
plt.title('Distribución del Período Orbital de Planetas')
plt.xlabel('Período Orbital (días)')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot para el dato cuantitativo 'orbital_period'
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='orbital_period')
plt.title('Boxplot del Período Orbital de Planetas')
plt.xlabel('Período Orbital (días)')
plt.show()
```

**La interpretación de tu gráfico anterior:** ------
La gráfica muestra que los planetas tienen un periodo orbital casi igual, eso significa que la diferencia entre planteas(masa) no es tan abismal, se tiene un rango de planetas que tienen un parecido periodos de orbita. se peude llevar por decenas de peso, pero no por miles de peso.

### **Ejercicio 3**: Método gráfico multivariado para datos cuantitativos


```python
#Tu código aquí
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset('planets')

# Crear un scatter plot para 'orbital_period' y 'mass', con el tamaño de los puntos representando 'distance'
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='orbital_period', y='mass', size='distance', hue='distance', palette='viridis', sizes=(20, 200), legend='brief')
plt.title('Período Orbital vs Masa de Planetas con la Distancia Representada por el Tamaño y Color')
plt.xlabel('Período Orbital (días)')
plt.ylabel('Masa (jupiter masses)')
plt.legend(title='Distancia (pc)')
plt.show()
```

**La interpretación de tu gráfico anterior:** ------
Tener una masa mas pesada tiende a demorar en su periodo orbital, planetas menos pesados, tiene una velocidad menor por la inercia de giro(masa)
